## Python Loop and Gradient Descent

```python
updated_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [updated_x]
slope_list = [previous_x]

for step in range(50):
    previous_x = updated_x
    gradient = df(previous_x)
    updated_x = previous_x - step_multiplier * gradient
    
    x_list.append(updated_x)
    slope_list.append(df(updated_x))
    
    step_size = abs(updated_x - previous_x)
    if step_size < precision:
        print("Loop run", step, "times.")
        break
    
    
print("Local minumum occurs at:", previous_x)
print("Slope of df(x) value at this point is:", df(updated_x))
print("f(x) cost at this point is:", f(updated_x))
```
