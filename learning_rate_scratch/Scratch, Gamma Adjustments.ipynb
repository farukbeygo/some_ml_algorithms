# Gamma Adjustments

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
sys.float_info.max
```

```python
# Creating our data
x = np.linspace(start=-5, stop=5, num=100)
```

```python
# defining our functions
def f(x):
    return x**4 - 4*x**2 + 5

def df(x):
    return 4*x**3 - 8*x
```

```python
def gradient_descent_function(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                             max_iter=300):
    updated_x = initial_guess
    list_of_x = [updated_x]
    slope_of_x = [derivative_func(updated_x)]
    
    for step in range(max_iter):
        previous_x = updated_x
        gradient = derivative_func(previous_x)
        updated_x = previous_x - multiplier * gradient
        
        step_size = abs(updated_x - previous_x)
        
        list_of_x.append(updated_x)
        slope_of_x.append(derivative_func(updated_x))
        
        if step_size < precision:
            break
    
    return updated_x, list_of_x, slope_of_x
```

```python
# Low rate of gamma
n = 100
low_gamma = gradient_descent_function(derivative_func=df, initial_guess=3, multiplier=0.0005, precision=0.00001, max_iter=n)

# Plotting the learning rate
plt.figure(figsize=[20, 10])
plt.xlim(0, n)
plt.ylim(0, 50)

plt.title('Learning rate at low gamma', fontsize=17)
plt.xlabel('Number of iteration', fontsize=15)
plt.ylabel('Cost rate', fontsize=15)

# Data of y axis
low_values = np.array(low_gamma[1])

# Data of x axis
iteration_list = list(range(0, n+1))

plt.plot(iteration_list, f(low_values), color='green', linewidth=2)
plt.scatter(iteration_list, f(low_values), color='green')

plt.show()
```

```python
# Big rate of gamma
n = 24
big_gamma = gradient_descent_function(derivative_func=df, initial_guess=3, multiplier=0.025, precision=0.00001, max_iter=n)

# Plotting the learning rate
plt.figure(figsize=[20, 10])
plt.xlim(0, n)
plt.ylim(0, 50)

plt.title('Learning rate at large gamma', fontsize=17)
plt.xlabel('Number of iteration', fontsize=15)
plt.ylabel('Cost rate', fontsize=15)

# Data of y axis
big_values = np.array(big_gamma[1])

# Data of x axis
iteration_list = list(range(0, n+1))

plt.plot(iteration_list, f(big_values), color='red', linewidth=2)
plt.scatter(iteration_list, f(big_values), color='red')

plt.show()
```

```python
# Smallest of gamma
n = 100
smallest_gamma = gradient_descent_function(derivative_func=df, initial_guess=3, multiplier=0.1e-4, precision=0.00001, max_iter=n)

# Plotting the learning rate
plt.figure(figsize=[20, 10])
plt.xlim(0, n)
plt.ylim(40, 60)

plt.title('Learning rate at smallest gamma', fontsize=17)
plt.xlabel('Number of iteration', fontsize=15)
plt.ylabel('Cost rate', fontsize=15)

# Data of y axis
smallest_values = np.array(smallest_gamma[1])

# Data of x axis
iteration_list = list(range(0, n+1))

plt.plot(iteration_list, f(smallest_values), color='blue', linewidth=2)
plt.scatter(iteration_list, f(smallest_values), color='blue')

plt.show()
```
